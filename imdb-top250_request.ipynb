**IMDB TOP 250**

```python
imdbUrl = 'https://www.imdb.com/chart/top/'

response = requests.get(imdbUrl)

soup = BeautifulSoup(response.content, 'html.parser')
```

```python
response_data = soup.find_all('table',{'class': 'chart full-width'}) #table bloÄunun chart full widht classÄ±nÄ± buluyoruz
response_data
```

```python
len(response_data[0].contents)
```

```python
response_data[0].contents #filmler tbody iÃ§inde, tbody 3.indexte gibi fakat \n'ler de olduÄu iÃ§in 5.indexte buldum
```

```python
movie_table = (response_data[0].contents)[5] #tbodyi movie tablea attÄ±m
movie_table
```

```python
movie_tableTr = movie_table.find_all('tr') #filmler tr taglerinin iÃ§inde
movie_tableTr
```

```python
# tr iÃ§indeki td taglerinde title colum clasÄ± var onlara ulaÅÄ±rsam film isimlerini almÄ±Å olacaÄÄ±m.
name = []
for movie in movie_tableTr:
    movie_title = movie.find_all('td',{'class': 'titleColumn'}) #250 filmin title bilgisini aldÄ±m
    movie_name = movie_title[0].text #textleri isimleri
    movie_name = movie_name.replace('\n','')#daÄÄ±nÄ±k duruyordu \nleri boÅlukla deÄiÅtirdim
    print(movie_name) 
    name.append(movie_name)
```

```python
rating = []
for movie in movie_tableTr:#rating bilgilerini Ã§ektim
    movie_title = movie.find_all('td',{'class': 'ratingColumn imdbRating'})
    movie_rating = movie_title[0].text
    movie_rating = movie_rating.replace('\n','')
    print(movie_rating)
    rating.append(movie_rating)
```

```python
rating = pd.Series(rating)
name = pd.Series(name)

df = pd.DataFrame(name+rating)
```
